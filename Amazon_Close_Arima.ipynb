```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd

# Load in the time series
amzn_df = pd.read_csv('amazon_close.csv', 
            index_col='date',
            parse_dates=True)

# Plot and show the time series on axis ax1
fig, ax1 = plt.subplots()
amzn_df.plot(ax=ax1)
plt.title('Amazon Stock Price from 2014-2019')
plt.show()
```

```python
#sort index
amzn_df = amzn_df.sort_index()
```

```python
# Split the data into a train and test set
amzn_train = amzn_df.loc[:'2018-01-01':]
amzn_test = amzn_df.loc['2018-01-01':]

# Create an axis
fig, ax = plt.subplots()

# Plot the train and test sets on the axis ax
amzn_train.plot(ax=ax)
amzn_test.plot(ax=ax)
plt.show()
```

<h2>Test if time series training data is stationary</h2>
<p>- Use Dickey Fuller Test</p>

```python
from statsmodels.tsa.stattools import adfuller

# Run the ADF test on the time series
result = adfuller(amzn_train['close'])

# Plot the time series
fig, ax = plt.subplots()
amzn_train.plot(ax=ax)
plt.show()

# Print the test statistic and the p-value
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

<h2>Since time series training data is not stationary, need to difference to get it stationary</h2>

```python
# Calculate the first difference of the time series
amzn_train_stationary = amzn_train.diff().dropna()

# Run ADF test on the differenced time series
result = adfuller(amzn_train_stationary['close'])

# Plot the differenced time series
fig, ax = plt.subplots()
amzn_train_stationary.plot(ax=ax)
plt.show()

# Print the test statistic and the p-value
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

<h2>P-Value is less than .05, thus, we can conclude time series is now stationary. We can forecast the data now using ARIMA!</h2>

```python
#Basic ARIMA model trying model order (1,1,1)

# Import the ARIMA model
from statsmodels.tsa.arima.model import ARIMA

# Instantiate the model
model = ARIMA(amzn_train['close'], order=(1,1,1))

# Fit the model
results = model.fit()

#see results
print(results.summary())
```

<h2>Make one-step-ahead predictions</h2>
<p>At any point in the time series, we can estimate the next series value.</p>
<p>Make this prediction by multiplying the previous value by the lag 1 AR coefficient</p>
<p>If previous value was 10 and AR coefficient was .6, then estimate next value is 6</p>
<p>If shock term had a standard deviation of 1, then predict our lower and upper uncertainty limits to be 5 and 7</p>
<p>&nbsp;</p>
<p></p>

```python
#make in-sample prediction
#-25= make predictions for last 25 entries of training data
forecast = results.get_prediction(start=-25)
#the central value is stored in the predicted_mean
mean_forecast = forecast.predicted_mean

#get confidence intervals of forecasts
confidence_intervals = forecast.conf_int()

# Select lower and upper confidence limits
lower_limits = confidence_intervals.loc[:,'lower close']
upper_limits = confidence_intervals.loc[:,'upper close']

print(mean_forecast)
```

```python
plt.figure()

#plot prediction
plt.plot(amzn_train.index[-25:],
         mean_forecast.values,
         color='red',
         label='forecast')

plt.fill_between(amzn_train.index[-25:], lower_limits, upper_limits, color='pink')
```

<h2>Reconstructing original time series after differencing</h2>
<p>Say you have to difference your time series in order to get the time series to be stationary and use ARIMA.</p>
<p>Except once you difference your dataset and use ARIMA, your forecast will be based on the differenced data</p>
<p>You'll need to convert your time series and forecast back to it's original format by taking the cumulative sum.</p>
<p>&nbsp;</p>
<p>&nbsp;</p>

```python
#DONT USE THIS ONE, CONFIDENCE INTERVALS ARE MESSED UP TRYING TO BACK TRACK DIFFERENCED DATA
steps = 50

amzn_train_stationary = amzn_train.diff().dropna()

# Instantiate the model
model = ARIMA(amzn_train_stationary['close'], order=(1,1,1))

# Fit the model
results = model.fit()

diff_forecast = results.get_forecast(steps=steps).predicted_mean

#convert differenced forecast back to forecast
from numpy import cumsum
mean_forecast = cumsum(diff_forecast) + amzn_train.iloc[-1,0]

# Calculate the confidence intervals
conf_int = results.get_forecast(steps=steps).conf_int()

```

<h2>Searching over Model order for optimal AIC and BIC</h2>
<p>- BIC favors simpler models than AIC</p>
<p>- AIC is better at choosing predictive models</p>
<p>- BIC is better at choosing good explanatory models</p>

```python
# Create empty list to store search results
order_aic_bic=[]

# Loop over p values from 0-2
for p in range(0,3):
  # Loop over q values from 0-2
    for q in range(0,3):
      	# create and fit ARMA(p,q) model
        # Instantiate the model
        model = ARIMA(amzn_train['close'], order=(p,1,q))
        results = model.fit()
        
        # Append order and results tuple
        order_aic_bic.append((p,q, results.aic, results.bic))
        
```

```python
# Construct DataFrame from order_aic_bic
order_df = pd.DataFrame(order_aic_bic, 
                        columns=['p', 'q', 'AIC', 'BIC'])

# Print order_df in order of increasing AIC
print(order_df.sort_values(by='AIC'))

# Print order_df in order of increasing BIC
print(order_df.sort_values(by='BIC'))
```

```python
#for a good model:
#residuals plot: no obvious pattern in residuals, should look like white nosie
#histrogram plus estimated density: two lines should almost be the same (follow normal curve)
#normal q-q plot: all points should lie along the red line, except for some values at the ends
#correlogram: 95% for lag greater than 0 should be within the blue shaded area (not significant). If there is significant correlation in residuals, then information that model hasn't captured
results.plot_diagnostics()
plt.show()
```

```python
#prob Q (Ljung Box Test): p value for null hypothesis that there are no correlations in the residuals (less than .05: correlated, greater than .05: not correlated)
#prob JB: p value for null hypothesis that residuals are normal(less than .05, not normally distributed, greater than .05, normally distributed)

#NOTE#
#If either p-value is less than 0.05, we reject that hypothesis
#In this case, they are NOT correlated but residuals are NOT normally distributed
#IDEALLY, BOTH PROB(Q) AND PROB(JB) SHOULD BE GREATER THAN .05 for a good model

#In this case, you could go back and apply some transformation to remove this outlier or you probably just continue to the production stage.
#could remove outliers and connect data points instead to smooth out data. Then re-apply model

print(results.summary())
```

<h2>With monthly data, we take first difference and seasonal difference (monthly=12) to make it stationary</h2>

```python
# Run ADF test on the differenced time series
result = adfuller(amzn_train_month_diff['close'])

# Plot the differenced time series
fig, ax = plt.subplots()
amzn_train_month_diff.plot(ax=ax)
plt.show()

# Print the test statistic and the p-value
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

<h2 id="After-differencing,-use-ACF-and-PACF-to-determine-non-seasonal-and-seasonal-parameters">After differencing, use ACF and PACF to determine non seasonal and seasonal parameters</h2>
<h2>Non-Seasonal Parameters</h2>

```python

# Import
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Create the figure 
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))

# Plot the ACF on ax1
plot_acf(amzn_train_month_diff, lags=11, zero=False, ax=ax1)

# Plot the PACF on ax2
plot_pacf(amzn_train_month_diff, lags=11, zero=False, ax=ax2)

plt.show()
```

```python
# Make list of lags
#can add more, but out of dates due to differencing
lags = [12]

# Create the figure 
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))

# Plot the ACF on ax1
plot_acf(amzn_train_month_diff, lags=lags, zero=False, ax=ax1)

# Plot the PACF on ax2
plot_pacf(amzn_train_month_diff, lags=lags, zero=False, ax=ax2)

plt.show()
```
